> **Tratando Valores Faltantes**



```python
#Consultando linhas com valores nulos
df.isnull().sum()
```

```python
#Substituindo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
#Verificar se ainda existe valores nulos
df.isnull().sum()
```

```python
#Verificando uma amostragem de dados
df.sample(10)
```

```python
#Apagando as linhas com valores nulos
df.dropna(inplace=True)
```

```python
#Apagando as linhas com valores nulos com base apenas em uma coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#Removendo linhas que estejam com os valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

> **Criando colunas novas**



```python
#Criando a coluna de Receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
#Criando o dataFrame
df = pd.read_excel("AdventureWorks.xlsx")
```

```python
#Quanitade de linhas e colunas
df.shape
```

```python
#Qual a receita total
df["Valor Venda"].sum()
```

```python
#Qual o custo total - Criando coluna para retornar essa informação
df["Custo"] = df["Custo Unitário"].mul(df["Quantidade"])
```

```python
#Amostragrem de 2 primeiras linhas
df.head(2)
```

```python
#Qual o valor do custo total
round(df["Custo"].sum(), 2)
```

```python
#Criando coluna para informar o Lucro (Receita - Custo)
df["Lucro"] = df["Valor Venda"] - df["Custo"]
```

```python
#Amostragem para exibir coluna nova
df.head(1)
```

```python
#Total  lucro
round(df["Lucro"].sum(), 2)
```

```python
#Criando coluna com o total de dias para enviar o produto (Data Envio - Data Venda)
df["Tempo Envio"] = df["Data Envio"] - df["Data Venda"]
```

```python
#Listagem para exibição da nova coluna
df.head(1)
```

```python
#Extrair apenas os dias
df["Tempo Envio"] = (df["Data Envio"] - df["Data Venda"]).dt.days
```

```python
#Verificando o tipo da coluna Tempo De Envio
df["Tempo Envio"].dtypes
```

```python
#Média do tempo de envio por Marca
df.groupby("Marca")["Tempo Envio"].mean()
```

```python
#Verificando se existe dados faltantes
df.isnull().sum()
```

```python
#Agrupar por ano e marca
df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum()
```

```python
#Resetando o Index
lucro_ano = df.groupby([df["Data Venda"].dt.year, "Marca"])["Lucro"].sum().reset_index()
lucro_ano
```

```python
#Qual o total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False)
```

```python
#Gráfico total de produtos vendidos
df.groupby("Produto")["Quantidade"].sum().sort_values(ascending=True).plot.barh(title="Total Produtos Vendidos")
plt.xlabel("Total")
plt.xlabel("Produto");
```

```python
#Gráfico Total de Lucro das Vendas por Ano
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum().plot.bar(title="Lucro x Ano")
plt.xlabel("Ano")
plt.xlabel("Receita");
```

```python
#Agrupamento das informações acima
df.groupby(df["Data Venda"].dt.year)["Lucro"].sum()
```

```python
#Selecionando apenas as vendas do ano de 2009
df_2009 = df[df["Data Venda"].dt.year == 2009]
```

```python
#Exibindo as informações da variavel
df_2009.head(2)
```

```python
#Gráfico em linhas das vendas de 2009 (Lucro x mês)
df_2009.groupby(df_2009["Data Venda"].dt.month)["Lucro"].sum().plot(title="Lucro x Mês")
plt.xlabel("Mês")
plt.xlabel("Lucro");
```

```python
#Gráfico em barras horizontal para lucro por Marcas das vendas de 2009 (Lucro x mês)
df_2009.groupby("Marca")["Lucro"].sum().plot.bar(title="Lucro x Marca")
plt.xlabel("Marca")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```

```python
#Gráfico em barras horizontal para lucro por Marcas das vendas de 2009 (Lucro x Classe)
df_2009.groupby("Classe")["Lucro"].sum().plot.bar(title="Lucro x Classe", color="lightgreen")
plt.xlabel("Classe")
plt.ylabel("Lucro")
plt.xticks(rotation='horizontal');
```
